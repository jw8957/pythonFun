```python
import PIL
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

```python
print('Pillow Version:', PIL.__version__)
```

```python
imgObj = PIL.Image.open('File/lena512.bmp')
```

```python
print(imgObj.format, imgObj.size, imgObj.mode)
```

```python
display(imgObj)
# plt.imshow(imgObj)
```

```python
imgArr = np.array(imgObj)

print(imgArr.dtype)
# uint8

print(imgArr.ndim)
print(imgArr.shape)
```

```python
print(imgArr)
```

```python
newImg = Image.fromarray(imgArr)
```

```python
print(newImg)
```
